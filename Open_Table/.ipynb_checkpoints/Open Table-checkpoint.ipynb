```python
import numpy as np
import pandas as pd
import plotly
from plotly.graph_objs import Figure, Histogram, Layout, Scatter, Bar
import scipy as stats
plotly.offline.init_notebook_mode(connected=True)
```

```python
# Correlation, groupby (location, cuisine, price), bar charts, 
```

```python
opentable = pd.read_csv('opentable_data.csv', index_col = 0) #import
opentable['num_reviews'] = opentable['num_reviews'].replace(np.nan, '0') #replace 'naN' with 0 reviews
opentable['recommendation_percentage'] = opentable['recommendation_percentage'].apply(lambda x: x.replace('%','')).astype(int) #make recommendation percentage an integer for analysis
opentable['num_reviews'] = opentable['num_reviews'].apply(lambda x: int(x)) # make number of reviews an integer (cleaning)
opentable = opentable.drop_duplicates() # delete duplicate rows
```

```python
location = opentable.groupby('location') #group by locations
```

```python
location.mean().sort_values('food_rating', ascending = False) #different analysis here (mean, var, count)
```

```python
cuisine = opentable.groupby('cuisine') #group by cuisine
```

```python
cuisine.mean().sort_values('food_rating', ascending = False) #different analysis here (mean, var, count)
```

```python
price = opentable.groupby('price')
```

```python
price.mean().sort_values('food_rating', ascending = False)
```

```python
has_chef_name = opentable['chef'].map(lambda x: (True if type(x)!=float else False))
```

```python
opentable['has_chef_name'] = has_chef_name
```

```python
chef = opentable.groupby('has_chef_name')
chef.mean().sort_values('food_rating', ascending=False)
```
